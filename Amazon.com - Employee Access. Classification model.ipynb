```python
#storing all action 1 rows in a df
action_1 = dataset.groupby('ACTION').get_group(1)
```

```python
action_1
```

```python
#sampling action 1 rows
sampled_action_1 = action_1.sample(1897)
```

```python
#printing action 1 sampled data
sampled_action_1.shape
```

```python
#Joining both datasets action_0 and sampled_action_1
new_dataset = action_0.append(sampled_action_1,sort=True)
len(new_dataset)
```

```python
#mixing up the rows
new_dataset = new_dataset.sample(len(new_dataset))
new_dataset
```

```python
#extracting features
X_feature = new_dataset.drop('ACTION',axis=1).values
```

```python
Y_target = new_dataset['ACTION'].values
```

```python
#Importing Modules for Prediction
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
logReg = LogisticRegression()
sc = StandardScaler()
```

```python
#spliting training and test data
X_train,X_test,Y_train,Y_test = train_test_split(X_feature,Y_target)
```

```python
#fitting the standard scaler module
sc.fit(X_train)
```

```python
#transforming training data into stander scaler
X_train = sc.transform(X_train)
```

```python
#training the logistic regression model
logReg.fit(X_train,Y_train)
```

```python
#making prediction from the model
Y_predict = logReg.predict(sc.transform(X_test))
```

```python
#Eveluating ther Logistic Regression model
from sklearn.metrics import accuracy_score,average_precision_score,recall_score,confusion_matrix,f1_score,classification_report
```

```python
print('Logistic Regression Model with Standardization')
print('Accuracy Score is     ',accuracy_score(Y_test,Y_predict))
print('Precision Score is    ',average_precision_score(Y_test,Y_predict))
print('Recall Score is       ',recall_score(Y_test,Y_predict))
print('Confusion Metrix is   ')
print(confusion_matrix(Y_test,Y_predict))
print('F1 Score is           ',f1_score(Y_test,Y_predict))
print (classification_report(Y_test,Y_predict))
```

```python
#using knn classifier
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
#fitting knn model
knn.fit(X_train,Y_train)
```

```python
#making prediction from the KNNeighbors model
Y_predict = knn.predict(sc.transform(X_test))
```

```python
print('KN Neighbors Model with Standardization')
print('Accuracy Score is     ',accuracy_score(Y_test,Y_predict))
print('Precision Score is    ',average_precision_score(Y_test,Y_predict))
print('Recall Score is       ',recall_score(Y_test,Y_predict))
print('Confusion Metrix is   ')
print(confusion_matrix(Y_test,Y_predict))
print('F1 Score is           ',f1_score(Y_test,Y_predict))
print (classification_report(Y_test,Y_predict))
```
